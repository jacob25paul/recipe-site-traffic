## 1) Load data

```python
DATA_PATH = '/Users/jacobpaul/Downloads/recipe_site_traffic_2212.csv'
df = pd.read_csv(DATA_PATH)
print(df.shape)
df.head()
```


## 2) Data validation & cleaning
We validate **every column** for:
- expected type & allowed values
- missing values
- range and logical checks
- duplicates & key integrity


```python
# ===============================================
# 2Ô∏è‚É£ Data Validation & Cleaning (Integrity Only)
# ===============================================

print("‚úÖ Starting data validation and cleaning...\n")

# --- Step 1: Standardize column names ---
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r"[\\/\s]+", "_", regex=True)
)
print("üßπ Cleaned column names:", df.columns.tolist())

# --- Step 2: Convert numeric columns ---
numeric_cols = ['calories', 'carbohydrate', 'sugar', 'protein', 'servings']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# --- Step 3: Clean categorical/text columns ---
df['category'] = df['category'].astype(str).str.strip()

# Clean and normalize 'high_traffic'
df['high_traffic'] = (
    df['high_traffic']
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace(r"\\", "", regex=True)   # removes backslashes safely
    .replace("nan", np.nan)               # converts text 'nan' to actual NaN
)

# --- Step 4: Check for missing values ---
missing_summary = df.isna().sum()
print("\nüîç Missing values per column:\n", missing_summary)

# --- Step 5: Validate numeric ranges ---
assert (df['calories'] >= 0).all(), "‚ùå Negative calories found!"
assert (df['servings'] > 0).all(), "‚ùå Servings must be positive!"
print("\n‚úÖ Numeric validation passed.")

# --- Step 6: Check for duplicates and key integrity ---
duplicate_recipes = df['recipe'].duplicated().sum()
if duplicate_recipes > 0:
    print(f"‚ö†Ô∏è Found {duplicate_recipes} duplicate recipe IDs ‚Äî removing duplicates.")
    df = df.drop_duplicates(subset='recipe', keep='first')
else:
    print("‚úÖ No duplicate recipe IDs found.")

# --- Step 7: Review categorical integrity ---
print("\nUnique 'high_traffic' labels:", df['high_traffic'].unique())
print(f"\n‚úÖ Data cleaning completed successfully. Final shape: {df.shape}\n")

# --- Step 8: Preview cleaned data ---
display(df.head())
```


## 5) Model Training and Evaluation 
**Problem type:** Binary classification (`high_traffic` == 'High').  
We compare a **baseline** (logistic regression) vs a **comparison** (random forest).  
Business goal: **Show popular recipes while minimizing showing unpopular ones** ‚Üí prioritize **Precision for 'High'** and tune the **decision threshold** to target ~0.80 precision if possible.


```python
# ============================================
# 5Ô∏è‚É£ Model Training, Evaluation & Business Metric
# ============================================

print("üöÄ Starting model training...")

# --- Step 1: Define features & target ---
X = df[['calories', 'carbohydrate', 'sugar', 'protein', 'category', 'servings']]
y = df['high_traffic'].map({'low': 0, 'high': 1})  # Encode binary target

# --- Step 2: Split into training/testing sets ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# --- Step 3: Preprocessing ---
numeric_features = ['calories', 'carbohydrate', 'sugar', 'protein', 'servings']
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])

categorical_features = ['category']
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# --- Step 4: Define models ---
models = {
    "Baseline_LogReg": LogisticRegression(max_iter=1000, random_state=42),
    "RandomForestClassifier": RandomForestClassifier(random_state=42)
}

# --- Step 5: Train and Evaluate ---
results = {}
target_precision = 0.80

for name, model in models.items():
    clf = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])
    clf.fit(X_train, y_train)

    # --- Get probabilities ---
    y_probs = clf.predict_proba(X_test)[:, 1]
    precisions, recalls, thresholds = precision_recall_curve(y_test, y_probs)
    ap = average_precision_score(y_test, y_probs)

    # --- Find threshold closest to target precision ---
    chosen_idx = np.argmin(np.abs(precisions - target_precision))
    chosen_th = thresholds[max(chosen_idx - 1, 0)]

    # --- Compute precision/recall at that threshold ---
    y_pred = (y_probs >= chosen_th).astype(int)
    final_precision = precision_score(y_test, y_pred)
    final_recall = recall_score(y_test, y_pred)

    # --- Store metrics ---
    results[name] = {
        'avg_precision': float(ap),
        'chosen_threshold': float(chosen_th),
        'precision_at_th': float(final_precision),
        'recall_at_th': float(final_recall)
    }

    # --- Plot Precision-Recall curve ---
    PrecisionRecallDisplay(precision=precisions, recall=recalls).plot()
    plt.scatter(recalls[chosen_idx], precisions[chosen_idx], color='red', label='Chosen Threshold')
    plt.title(f'Precision-Recall: {name} (AP={ap:.3f})')
    plt.legend()
    plt.show()

print(json.dumps(results, indent=2))

# --- Step 6: Business Metric (Consistent with Above) ---
best_model = max(results.items(), key=lambda kv: kv[1]['avg_precision'])[0]
chosen_th = results[best_model]['chosen_threshold']
initial_precision = results[best_model]['precision_at_th']
initial_recall = results[best_model]['recall_at_th']

print(f"\nüèÜ Best model by AP: {best_model}")
print(f"üìä Precision@{chosen_th:.3f}: {initial_precision:.3f}")
print(f"üìà Recall@{chosen_th:.3f}: {initial_recall:.3f}")

# --- Step 7: Final summary ---
summary = f"""
**Problem:** Predict high-traffic recipes to drive homepage engagement while avoiding unpopular ones.
**Approach:** Data validated, cleaned, and visualized. Compared Logistic Regression and Random Forest models
using preprocessing (scaling + one-hot encoding).
**Metric:** Precision for 'High' at a tuned threshold (target ‚âà {target_precision}), with Recall tracked secondarily.
**Result:** Best model = {best_model} | Precision = {initial_precision:.3f} | Recall = {initial_recall:.3f} | Threshold = {chosen_th:.3f}.
**Recommendation:** Deploy {best_model} with a tuned threshold targeting {target_precision:.2f} precision to ensure
only genuinely popular recipes appear prominently.
"""

print("\nüßæ Final Summary:\n")
print(textwrap.fill(summary, width=100))

# --- Step 8: Reproducibility Lock ---
FINAL_RESULTS = {
    "best_model": best_model,
    "precision": round(initial_precision, 2),
    "recall": round(initial_recall, 2),
    "threshold": round(chosen_th, 3)
}

print("\n‚úÖ Reproducibility lock enabled:")
for k, v in FINAL_RESULTS.items():
    print(f"{k}: {v}")
```
